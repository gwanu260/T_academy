### Type2 프로젝트 생성 -> 원격 저장소 생성 및 연결

- 목표
  - 로컬 PC에서 개발한 프로젝트를 특정 레포지토리로 연동
  - 프로젝트 폴더명과 레포지토리명이 불일치하면 다른 PC에서 clone 할때 폴더명을 프로젝트명으로 지정해서 진행해야 함
  ```
    # 폴더명 불일치
    $ git clone git@github.com:ucoccto/sb_test.git test

    # 폴더명 일치
    $ git clone git@github.com:ucoccto/sb_test.git
  ```

- 연동 작업
  - ps(파워쉘) 가동
  - 해당 프로젝트 폴더로 이동
  ```
    PS C:\test>
  ```
  - 로컬 공간에 git 초기화 진행
  ```
    $ git init
  ```
    - 특정 폴더를 깃 저장소로 만들 때 사용하는 명령어
    - .git 폴더 생성됨(숨김 파일)
      - 코드의 변경 내역(버전 포함) 관리를 위한 정보 저장하는 공간
  
  - 깃헙 저장소와 로컬 깃 저장소 연결
    - remote 명령어
      - 로컬 깃 저장소의 모든 이력, 파일 -> 깃 허브에 업로드 하기 위해 연결하는 명령어
    - 깃헙에 새로운 레포지토리 생성
      - 이름은 test(가급적 프로젝트 폴더명과 동일하게) 입력
    - 연결
      - 로컬 저장소와 원격 저장소 연결(링크)
      ```
        $ git remote add origin git@github.com:ucoccto/test.git
      ```
    - 스테이지 추가
      - 대상 : 로컬 저장소 변경이력, 신규파일, 수정파일..
      - 해당 파일들과 변경내역을 추적하여, 스테이지라는 상태로 변경처리함(스테이지에 추가한다)
      ```
        $ git add .
      ```

    - 로컬저장소 업로드
      - 커밋 (commit)
      - 스테이지에 존재하는 내용(변경이력, 변경파일, 신규파일,..) 로컬 저장소에 올림(업로드)
      ```
        $ git commit -m "Init : springboot init"
      ```

    - 원격저장소 업로드
      - 브랜치명 main으로 적용
        - 깃에서는 브랜치별로 작업을 분할해서 진행 가능함
        - 브랜치 변경이 없으면 1회성
        ```
          $ git branch -M main
        ```
      - 푸시 : 원격 저장소에 로컬 저장소 내용을 동기화(업로드)
        - 깃헙에 반영
        - push <-> pull
        ```
          $ git push origin main
        ```

## 심플 구조

- GitAction Flow
  - https://docs.github.com/en/get-started/using-github/github-flow
  
  - 2종류 브런치만 존재
    - main : 언제든지 배포 가능한 상태 유지, main에 기타 브런치가 병합될때 엄격하게 검수
      - ci를 활용하면 좋음
    - 기타 브런치 : 작업 단위던, 자유롭게

    ```
      main
        L etc branch1
        L etc branch2
        L etc branch3
    ```

## 복합구조

- main or master :
  - 운영환경에 배포될 수 있는 코드를 모아둔다.

- develop :
  - 다음 버전 기능이 담긴 코드
  - 기능을 운영환경에 업데이트 할 때 main으로 merge한다.

- feature :
  - 새로운 기능을 개발하기 위한 브랜치
  - develop 브랜치에서 생성
  - 기능 개발이 완료되면 develop으로 merge한다.

- release :
  - 새로운 버전의 배포를 위한 브랜치로,
  - develop 브랜치에서 생성
  - 배포 전 버전 이름 등 사소한 데이터 수정 및 버그를 수정하기 위해 사용
  - 배포 준비가 완료되면 main, develop에 둘 다 merge한다.

- hotfix :
  - 운영환경에 빠르게 수정해야할 버그가 있을 때 main을 빠르게 변경하기 위해 사용
  - main 브랜치에서 생성하고 버그 수정이 끝나면 main, develop에 merge한다.

# 깃 명령어

## 명령어

명령어 정리

---

### 1. **기본 명령어**
- `git init`  
  새로운 Git 저장소를 초기화.  

- `git clone <URL>`  
  원격 저장소를 로컬로 복제.  

- `git status`  
  현재 작업 디렉토리의 상태를 확인.  

- `git add <파일명>`  
  변경 사항을 Staging Area에 추가.  
  - 모든 파일을 추가하려면: `git add .`  

- `git commit -m "메시지"`  
  Staging Area의 변경 사항을 로컬 저장소에 커밋.  

- `git log`  
  커밋 내역을 확인.
  - 빠져나가기 => q 입력

---

### 2. **브랜치 관련 명령어**
- `git branch`  
  현재 브랜치를 확인하거나 새 브랜치를 생성.  
  - 새 브랜치 생성: `git branch <브랜치명>`  

- `git checkout <브랜치명>`  
  특정 브랜치로 전환.  
  (최근에는 `git switch` 명령으로 대체 가능)  

- `git switch -c <브랜치명>`  
  새 브랜치를 생성하고 전환.  

- `git merge <브랜치명>`  
  다른 브랜치를 현재 브랜치에 병합.  

- `git branch -d <브랜치명>`  
  브랜치를 삭제.  

---

### 3. **원격 저장소 관련 명령어**
- `git remote add origin <URL>`  
  원격 저장소를 추가.  

- `git remote -v`  
  연결된 원격 저장소를 확인.  

- `git push origin <브랜치명>`  
  로컬 커밋을 원격 저장소에 업로드.  

- `git pull origin <브랜치명>`  
  원격 저장소의 변경 사항을 로컬로 가져와서 병합
  - (*)가져와서 병합하는데 하필 로컬에서 겹치는 파일에 수정을 가했다 - error 발생, 작업실패
    - 항상 기기가 변경되면, 최신으로 유지후 작업 진행


- `git fetch`  
  원격 저장소의 업데이트를 가져오지만 병합하지는 않음.  

---

### 4. **파일 관리 관련 명령어**
- `git rm <파일명>`  
  파일을 Git에서 삭제.  

- `git mv <기존경로> <새경로>`  
  파일 또는 디렉토리를 이동하거나 이름을 변경.  

---

### 5. **Reset 및 Revert**
- `git reset --soft <커밋ID>`  
  특정 커밋으로 돌아가되, 변경 사항은 Staging Area에 남김.  

- `git reset --mixed <커밋ID>`  
  특정 커밋으로 돌아가고 변경 사항은 작업 디렉토리에 남김.  

- `git reset --hard <커밋ID>`  
  특정 커밋으로 돌아가고 변경 사항을 모두 삭제.  

- `git revert <커밋ID>`  
  특정 커밋의 변경 사항을 되돌리는 새로운 커밋을 생성.  

---

### 6. **기타 유용한 명령어**
- `git stash`  
  현재 작업 중인 변경 사항을 임시로 저장.  

- `git stash pop`  
  stash에 저장한 변경 사항을 다시 적용.  

- `git tag <태그명>`  
  특정 커밋에 태그를 추가.  

- `git diff`  
  변경된 내용을 비교.  
  - Staging Area와 비교: `git diff --staged`  

- `git blame <파일명>`  
  파일의 각 줄이 누가, 언제 변경했는지 확인.

## 스테이징 => 원위치, 다시스테이징

원위치 처리

---

### **스테이징 영역에서 변경 사항 제거**
1. **스테이징 영역 전체 초기화**
   ```bash
   git reset
   ```
   - 변경 사항은 **워킹 디렉토리(Working Directory)**에 남아 있음.
   - 즉, 파일의 수정 사항은 그대로 유지되지만, 커밋 준비 상태는 해제됨.

2. **특정 파일만 스테이징 취소**
   ```bash
   git reset <파일명>
   ```
   - 스테이징 영역에서 특정 파일만 제거하고 싶을 때 사용.

---

### **스테이징 취소 + 변경 내용 삭제**
만약 **변경 내용 자체를 삭제**하고 원래 상태로 되돌리고 싶다면:

1. **전체 파일 변경 내용 취소**
   ```bash
   git restore .
   ```
   - 워킹 디렉토리에서 수정한 모든 파일의 변경 사항이 삭제.

2. **특정 파일만 변경 취소**
   ```bash
   git restore <파일명>
   ```

---

### **정리**
- 코드 수정 -> 저장
  - 스테이징
    - `git add .`
  - 언스테이징
    - **스테이징만 취소**:
    - `git reset`  
  - 수정내용 원복
    - **수정 내용 자체 취소**
    - `git restore .`

  - vs code에서 -버튼 눌어서 대체 가능
    - staged  `- 버튼`
    - changes `- 버튼`


## 커밋, 되돌리기

`git commit -m "Feat: pos 작업"`을 되돌리고 싶다면

---

### **1. 가장 최근 커밋을 단순히 되돌리고 싶을 때**  
커밋만 제거하고 변경 사항은 유지하려면:

```bash
git reset --soft HEAD~1
```
- 이 명령어는 가장 최근 커밋을 삭제하지만, **커밋된 변경 사항은 스테이징 영역**에 남김.
- 이후 수정하거나 다시 커밋할 수 있음.

---

### **2. 최근 커밋 + 스테이징 취소**
커밋과 스테이징 영역도 모두 초기화하려면:

```bash
git reset --mixed HEAD~1
```
- 이 경우 커밋은 제거되고, 변경 사항은 **워킹 디렉토리로 이동**됨.

---

### **3. 커밋 + 수정 사항 모두 취소**
최근 커밋과 변경된 파일 내용 모두를 원상태로 되돌리려면:

```bash
git reset --hard HEAD~1
```
- **경고**: 이 명령어는 최근 커밋과 변경된 내용까지 모두 삭제. 되돌릴 수 없으니 신중히 사용.

---

### **4. 커밋을 취소하고 새로운 커밋으로 되돌리기**
만약 되돌린 기록을 **명시적으로 남기고** 싶다면, `git revert`를 사용:

```bash
git revert HEAD
```
- 이 명령어는 "되돌리는 커밋"을 새로 생성하여 작업 내역을 보존함.
- 팀 협업 환경에서 주로 사용되며, 기존 커밋 기록을 보존해야 할 때 유용.

---

### 상황별로 어떤 명령어를 사용할지 선택하세요:
1. **변경 사항은 유지하고 커밋만 취소** → `git reset --soft HEAD~1`  
2. **스테이징도 취소하고 커밋만 삭제** → `git reset --mixed HEAD~1`  
3. **모든 것을 초기 상태로 되돌림** → `git reset --hard HEAD~1`  
4. **기록을 남기며 커밋 취소** → `git revert HEAD`
