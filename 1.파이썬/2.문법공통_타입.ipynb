### 인덱싱

- 문자열에서 특정 문자열을 **추출**하는 행위
- 대부분 컬렉션 계열 자료구조에서 지원
- 본질적 의미
  - **차원축소**
    - 2차원 -> 인덱싱 -> 1차원
    - 리스트, 딕셔너리, 튜플, 배열, 시리즈, 데이터프레임, 텐서
  - 문자열은 차원축소 개념 미적응.
- 문법
  - 정방향 인덱스값
    - 문자열 앞에서부터 카운트
      - 왼쪽 -> 오른쪽
    - 0, 1, 2, ...

  ```
    변수명[ 정(or 역)방향 인덱스값 ]
  ```

  - 역방향 인덱스값
    - 문자열 뒤에서부터 카운트
      - 오른쪽 -> 왼쪽
  - -1, -2, -3, ...
  ```
    변수명[ 정(or 역)방향 인덱스값 ]
  ```

````python
# 더미 데이터
a = 'abcdefghijklmn'

a
````

````python
# a라는 변수가 참조하고 있는 문자열에서 1번자리(인덱스) 데이터를 추출하시오 -> 'b'
a[ 1 ]

# 실습 -> 역방향을 계산하여 동일한 결과를 추출하시오.
# 커서는 문자열과 문자열 사이 => 인덱스 자리임
# 인덱스 위치에 오른쪽값을 취하는게 기본형태
a[ -13 ]

# 값에 가까운쪽에서 정 OR 역 방향을 정하여 추출
````

### 슬라이싱 (자른다, 범위)

- 원본 데이터에서 특정 데이터만 자른다
- **차원 유지**
  - 문자열을 제외한 다른 자료구조에서 해당
    - 2D -> 슬라이싱 -> 2D
- 데이터는 연속적으로 이웃해야한다.
  - step 옵션을 통해서 간격을 조절할수 있다.
- 문법
  ```
    변수명 [ 시작인덱스 : 끝인덱스 : step(간격, 기본값 1, 생략가능함) ]

    or
    # 이 표현은 무조건 간격 1임
    변수명 [ 시작인덱스 : 끝인덱스]

    or
    # 시작인덱스 ~ 끝까지
    변수명 [ 시작인덱스 : ]

    or
    # 처음부터 ~끝까지 -> copy
    변수명 [ : ]
  ```

````python
a
````

````python
# 카피
a [ : ]
````

````python
# 처음부터 2개만 자른다, 앞쪽(prefix)에서 특정문자열 자를때 유용
# 0 <= 인덱스 < 2
a[ : 2]
````

````python
# 맨뒤에서 2개만 자른다.
# -2 <= 인덱스 < 끝까지
a[ -2: ]
````

````python
# 앞으로 한문자씩 제외한 것과 동일
a[ 1 : -1 ], a[ 1 : -1 : 1]
````

````python
# 간격 2 -> 하나씩 건너띠고 추출
a[ 1 : -1 : 2]
````

#### 공백제거

- 수집한 문자열, 사용자 입력 등등 불필요한 공백은 노이즈임.
- 문자열의 왼쪽, 오른쪽, 양쪽 공백 제거하는 함수 제공
- 복잡한 내부의 공백등은 -> 정규식적용, replace()함수 활용 가능
- strip(), lstrip(), rstrip()
  - 타 언어는 trim() 많이 보임

````python
# 더미 데이터
data = ' ab cd '

# 단순하게 처리하여 출력
print('[' + data +  ']') # 원본
print('[' + data.lstrip() +  ']') # 왼쪽 공백 제거
print('[' + data.rstrip() +  ']') # 오른쪽 공백 제거
print('[' + data.strip() +  ']') # 양쪽 공백 제거
````
