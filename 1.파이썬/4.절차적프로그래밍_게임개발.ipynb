# Stage1

- 요구사항
  - 게임을 시작하면, 다음과 같은 프롬포트 출력
  ```
    게임 제목을 입력해주세요. (최대 28자이내(<=28), 반드시 영문 or 숫자 만 사용가능)
  ```
  - 한글입력 X, 특수문자 X -> 검사 안함.
  - 사용자는 게임 제목을 입력후 엔터치면 완료
  - 사용자가 입력하고 엔터치기전까지 무한대기
  - 엔터후 체크 포인트
    - 입력 없이 엔터
    - 28자 이상 넣고 엔터
    - 공백만 넣고 엔터
    - 모두 오류메세지로 대응 : "정확하게 입력하세요" => 다시 입력대기
  - 정확하게 입력했다면
    - stage2로 진입
    - 게임 제목은 game_title이라는 변수로 세팅
    - 무한대기 종료
  
  - **제시되지 않은 조건은 가정하고 진행**

````python
# 환경변수와 같은 외부 리소스 요소 간주
START_PROMPT = '게임 제목을 입력해주세요. (최대 28자이내(<=28), 반드시 영문 or 숫자 만 사용가능)\n'
ERROR_MSG1 = '정확하게 입력하세요'
GAME_TITLE_MAX_LEN = 28

while True: # 무한루프
  game_title = input(START_PROMPT).strip()  # 사용자 입력 -> 공백제거 -> 변수세팅
  # print( game_title, len(game_title) ) # 테스트용, 향후제거

  if not game_title: # len(game_title) == 0: # 입력내용이없다면:
    # 오류메세지 => 다시 입력대기
    print(ERROR_MSG1)
  elif len(game_title) > GAME_TITLE_MAX_LEN: # 28자 이상이라면
    # 오류메세지 => 다시 입력대기
    print(ERROR_MSG1)
  else: # 정상입력
    # 입력대기 탈출 무한 루프 탈출(종료)
    break # 임시로 탈출 -> 1회반복
````

````python
while True:
  game_title = input(START_PROMPT).strip()
  if not game_title:
    print(ERROR_MSG1)
  elif len(game_title) > GAME_TITLE_MAX_LEN:
    # 요구사항 추가
    print(f'''
게임 제목을 정확하게 입력하세요.
최대 28자 이내로 입력하시고,
방금 입력한 내용은 {GAME_TITLE_MAX_LEN}이고,
최대 길이 대비 {len(game_title) - GAME_TITLE_MAX_LEN}자가 초과 되었습니다.
    ''')
  else:
    print('정상입력완료', game_title)
    break
````

````python
import random

#stage 1
while True:
  game_title = input(START_PROMPT).strip()
  if not game_title :
    print(ERROR_MSG1)
  elif len(game_title) > GAME_TITLE_MAX_LEN:
    # 요구사항 추가 -> "왜 않되는지 친철하게 안내라는 메세지 구성"
    print(f'''
게임 제목을 정확하게 입력하세요.
최대 {GAME_TITLE_MAX_LEN}자 이내로 입력하시고,
방금 입력한 내용은 `{game_title}`이고,
최대 길이 대비 { len(game_title) - GAME_TITLE_MAX_LEN }자가 초과 되었습니다.
    ''')
  else:
    print( '정상입력 완료', game_title)
    break

# stage 2
PLAYER_PROMPT = '플레이어의 이름을 입력해 주세요. (최대 28자이내(<=28), 반드시 영문 or 숫자 만 사용 가능)\n'
PLAYER_NAME_MAX_LEN = GAME_TITLE_MAX_LEN
while True:
  player_name = input(PLAYER_PROMPT).strip()
  if not player_name :
    print( ERROR_MSG1 )
    continue
  if len(player_name) > PLAYER_NAME_MAX_LEN:
    print( ERROR_MSG1 )
    continue

  # 정상입력을 수행해야만 진입하는 코드
  break

#stage 3
GAME_INTRO_ICON = '*'
GAME_INTRO_SPACE_SIZE = GAME_TITLE_MAX_LEN + 2 # 30
print( GAME_INTRO_ICON * GAME_INTRO_SPACE_SIZE )
print( f'{GAME_INTRO_ICON}{game_title:^28}{GAME_INTRO_ICON}')
print( f'{GAME_INTRO_ICON}{player_name:^28}{GAME_INTRO_ICON}')
print( f'{GAME_INTRO_ICON}{"press any key":^28}{GAME_INTRO_ICON}')
print( GAME_INTRO_ICON * GAME_INTRO_SPACE_SIZE )

#stage 7
is_game_playing = True
while is_game_playing:#조건문:
  # stage 4
  ai_number = random.randint(1, 100)
  print( 'ai_number', ai_number )
  try_count = 0

  # stage 6
  while True:
    # stage 5
    while True:
      player_number = input('AI가 생성한 값 1 ~ 100 사이중 하나의 정수값을 맞추시오\n').strip()
      if not player_number:
        print('값이 비워 있습니다. 정확하게 입력하세요.')
        continue
      if not player_number.isnumeric():
        print('정수값만 입력가능합니다. 정확하게 입력하세요.')
        continue
      player_number = int( player_number )
      if ( player_number < 1) or ( player_number > 100):
        print("정확하게 입력하세요.")
        continue
      break

    try_count += 1
    if player_number > ai_number:
      print( '힌트 - 크다')
      pass
    elif player_number < ai_number:
      print( '힌트 - 작다')
      pass
    else:
      print( '정답')
      try_count -= 1
      break

  print( f'{ (10-try_count)*10 }점 획득하였습니다.')

  # stage 7
  while True:
    player_answer = input('''
      다시 게임을 진행하시겠습니까?
      (yes|no or y|n)중 선택(대소문자 구분x)
      ''').strip().lower() # 내부적으로는 무조건 소문자로 처리
    if player_answer == 'yes' or player_answer == 'y':
      # stage 4부터 다시 진행
      break
    elif player_answer == 'no' or player_answer == 'n':
      # 종료 -> stage7 탈출 -> 중첩 반복문 탈출 -> 쉽게 구현:플레그 변수기법활용
      is_game_playing = False
      print('bye bye ~ ')
      break
    else:
      print( '정확하게 입력하세요.' )
````
