# 조건식

- 특징
  - 조건문, 반복문이 구동(실행)되게 하는 기본 전제 조건
- 결론
  - 조건식의 결론 -> True, False (조건|반목문 내부 진입 X) -> 불린형
    - False로 인식하는 표현
      - None, '', 0, 0.0, [], {}, (), False

- 연산자
  - 값 vs 값 비교하는 조건식
    ```
    A == B : A와 B는 (값이) 일치하는가?
    A != B : A와 B는 다른가?
    A > B : A는 B보다 큰가? => 초과
    A < B : A는 B보다 작은가? => 미만
    A >= B : A는 B보다 크거나 같은가? => 이상
    A <= B : A는 B보다 작거나 같은가? => 이하
    not A : A의 값을 부정한다
      - 부정상황을 체킹할때 사용
      - 부정 상황을 부정(not) -> True가 감지되면 -> A값은 비어있다(False상황이다)
    ```
  - 조건식 vs 조건식을 비교하는 연산자
    - 타언어에서는 쇼트연산자라고 부름
    ```
      # (조건식=c1) 연산자 (조건식=c2)
      # 조건이 복잡한 경우 빠른 처리를 위함, 불필요한 연산제거

      **C1 or C1**
        - **C1이 참이면 그대로 조건 체크 종료**
        - C2는 체크할 필요 없음. C2의 결과에 상관없이 무조건 참
        - C1이 거짓이면 C2까지 체크하고 결론내림
        - ~ 이거나 ~

      **C1 and C2**
        - **C1이 거짓이면 그대로 조건 체크 종료**
        - C2는 체크할 필요 없음. C2의 결과에 상관없이 무조건 종료
        - C1이 참이면 C2까지 체크하고 결론 내림
          - 참의 결론에 도달하기 위해서는 반드시 C1, C2rk ahencka
        - ~ 고 ~
    ```
  - 기타
    ```
    A is B
      - A는 B인가? => 객체를 관리하는 내부 id를 비교하는 연산자
      - 동일 객체인가? (본질이 같은가? 메모리상)

    a in A
      - A안에 멤버로 A가 존재하는가?
    ```

- 요구사항
  - pos기 기계의 결제라고 가정
  - 결제에 따른 메세지 전송
    - 조건
      - 체크카드(잔액은 5천원)잔액에 따라 아래 형태가 작동된다 가정
      - 총결제액이 5천원 초과 (>)
        - 메세지 : x원이 부족합니다.
      - 총결제액이 5천원 미만 (<)
        - 메세지 : 잔액 X원이 남았습니다.
      - 총결제액이 5천원 일치 (==)
        - 메세지 : 딱 맞게 결제하였습니다

````python
# 임시 설정
while True: # 반복문
  # input() : 내장함수, 사용자 입력 무한 대기 (블럭킹, 코드 진행 x)
  # 터미널, 표준입력, 콘솔등에서 입력대기
  # 코랩등에서는 웹환경이므로, 입력창으로 대체
  # 사용자가 입력후, 엔터키를 입력하면 종료됨

  # 주문 총액을 입력받음 -> 문자열 -> 공백제거 -> 정수변환
  coffee_price = int(input('구매 총액을 입력하세요\n').strip())
  print(coffee_price, type(coffee_price))

  # 반복문의 탈출코드
  # coffee_price의 값이 0과 같으면 break 수행
  if coffee_price == 0:
    # break를 감싸고 있는 모든 반복문들중, 가장 가까운 반복문탈출
    break
````

````python
# 고객의 체크카드의 잔액
customer_money = 5000

while True:
  coffee_price = int(input('구매 총액을 입력하세요\n').strip())
  if coffee_price == 0:
    break

  '''
    조건
      체크카드(잔액은 5천원)잔액에 따라 아래 형태가 작동된다 가정

      총결제액이 5천원 초과 (>) 라면
        메세지 : x원이 부족합니다.
      총결제액이 5천원 미만 (<) 이라면
        메세지 : 잔액 X원이 남았습니다.
      총결제액이 5천원 일치 (==) 한다면
        메세지 : 딱 맞게 결제하였습니다
  '''

  # 구매 총액에 따른 메시지 출력하시오(실습)
  if coffee_price > customer_money: # 조건식1:
    # 문자열 포멧팅을 통해서 동적으로 메세지 구성
    print(f'{coffee_price - customer_money}원이 부족합니다.')

  elif coffee_price < customer_money: # 조건식2:
    print(f'잔액 {customer_money - coffee_price}원이 남았습니다.')

  else: # coffee_price == customer_money
    print('딱 맞게 결제하였습니다.')
````

### 리스트 + 딕셔너리 형태의 반복문

````python
ktx_stations = [
  {
    'name':'영등포역',
    'age':70,
    'flag':[1,2,3]
  },
  {
    'name':'광명역',
    'age':25,
    'flag':[5,6,7]
  }
]
ktx_stations
````

````python
# 요구사항
'''
영등포역의 년식은 70년
광명역의 년식은 25년
'''
# 실습
for ktx_station in ktx_stations:
  # ['영등포역', 70, [1, 2, 3]] => 변수에 직접 할당
  name, age, _ = ktx_station.values()
  print(f'{name}의 년식은 {age}년')
  # break # 샘플 1개만 처리됨
for station in ktx_stations:
    print(station.items())
for station in ktx_stations:
    print(station.values())
````

````python
for ktx_station in ktx_stations:
  # 리스트나 튜플이 중복적 구조로 배치도이어 있으면 이중분해 가능함
  # (변수, 변수...) <= 표현
  name, age, (_, flag, _) = ktx_station.values()
  print(f'{name}의 년식은 {age}년, {flag}')
````

### range()

- 특정 범위내에 수를 생성(공급)하는 함수
- 제너레이터의 기능을 탑재하고 있음
  - 필요할때 필요한만큼 생성 -> 메모리를 효율적 관리 -> numpy에서 기본적으로 사용.
- 사용
  ```
    range( 시작값, 끝값, step(기본값1) )

    # 시작값 <= x < 끝값
    # 간격 기본 1
    # 시작값을 생략하면 0부터라는 의미
  ```

````python
# range( 3 ) => range ( 0, 3 )
# 주로 반복문하고 같이 사용
for x in range ( 3 ):
  print( x )

print( '-'*10 )
# 요구사항 1<= x < 3
# 실습
for x in range ( 1, 3 ):
  print( x )

print( '-'*10 )
# 요구사항 1 3 5 7 9 11 출력
for x in range ( 1, 12, 2 ):
  print(x)
````

### 반복문 + 조건문(필터링)

````python
# 짝수면 => 조건식
# 짝수 => 2로 나뉘어 떨어진다 => 나머지가 x => 나머지가 0이다
# % <= 나머지 연산자
3%2 # 홀수
4%2 # 짝수
````

````python
# 요구사항, 홀수값을 제외하고 출력하시오
# 단, range( 10 )은 수정 X
for n in range( 10 ): # 0 <= n < 10, STEP 1 단위로 생성 공급
  # 조건문 -> 짝수면 출력, 홀수면 패스(아무일도 하지 않음)
  # 실습 2분
  if n%2 == 0:
    print( n )
  else:
    pass
````

````python
# 아무것도 하지 않는 코드는 제거, 단 구조상 문제가 없다면
for n in range( 10 ):
  if n%2 == 0:
    print( n )
````

````python
# 조금더 파이썬의 불린 특성을 활용하여 표현
# 비교 연산자를 배제한 표
for n in range( 10 ):
  if not n % 2:
    print( n )
````

### 이중 for문

- for 문 안에 for문 존재
  - 중첩 반복문
    - 구구단
    - 게임의 타일 세팅
    - 딥러닝 CNN(영상인식에서 많이 사용)에서 필터가 2D 이미지를 탐색할 때(특성추출)

````python
# 구구단, 3단~6단
# 앞자리는 3, 4, 5, 6
# 두번째 자리 1,2,3, .., 9
# 세번째 자리 연산결과
'''
3 x 1 = 3
3 x 2 = 6
...
6 x 9 = 54
'''
for n in range( 3, 7):
  print( n )
  for m in range( 1, 10):
    print( n, m, n*m )


````

````python
# 포멧 적용
for n in range( 3, 7):
  for m in range( 1, 10):
    print(f'{n} x {m} = {n*m:>2}')
````

````python
# 5단은 제외하고 3단, 4단, 6단 출력하고,-> 필터링
# 3 x 3 제외, 4 x 4 제외, 6x6 제외 -> 필터링 -> 조건
# 실습 : 3분
for n in range( 3, 7):
  for m in range( 1, 10):
    if (n != m and n != 5):
      print(f'{n} x {m} = {n*m:>2}')
````

### break, continue, else

- break
  - 반복문을 중단(탈출)
  - **break를 감싸고 잇는 가장 가까운 반복문 탈출**

- continue
  - **continue를 감싸고 있는 가장 가까운 반복문으로 점프 -> 다음번 반복문 진행**

- else
  - for ~ else ~
  - while ~ else ~
  - 반복문이 중단없이 정상적으로 모두 반복행위를 하고 마무리되었다면 진입함. -> 강제로 반복문 종료되었는지 점검할 수 있다.
  - 필요하면 사용
  - 어떤의 변수 없이 그대로 반복문(혹은 예외처리)진행되면 진입 -> 뉘앙스

````python
# 목적
# 임시적 코드, 샘플링을 위해 사용, 1회성 작동
# 한번만  돌려보고 판단하기 위한 임시코드
for n in range( 10 ):
  print( n )
  break
````

````python
for n in range( 10 ):
  # n이 3을 초과하면, 중단
  if n > 3:
    break # break는 통상 조건문과 같이 사용된다. -> 탈출코드
  print( n )
````

````python
for n in range( 10 ):
  print( n )
  if n > 3:
    break # 중단위치는 요구사항에 맞게 잘 구성함.
````

````python
for n in range( 3, 7):
  for m in range( 1, 3):
    print(f'{n} x {m} = {n*m:>2}')
    break # 2번 라인의 for문과 반응하여 중단
  break # 1번 라인의 for문과 반응하여 중단
````

#### continue

- 반복문을 모두 수행은 한다.

````python
for n in range( 2, 6 ):
  # 짝수값을 제외하고 출력
  if not n % 2: # 짝수라면
    continue # 조건을 만족하면 아래코드 수행 X,  반복문(1번라인의 ragnfe())으로 점프
  print( n )
````

#### else

````python
for i in range(3):
  print( 'A', i )
  if i == 1:
    print( 'B', i )
    continue
  print( 'C', i )
else:
  print('cotinue 사용시 else에 진입한다')
````

````python
for i in range(3):
  print( 'A', i )
  if i == 1:
    print( 'B', i )
    break
  print( 'C', i )
else:
  print('break 사용시 else에 진입 안한다')
````

## While문

- 언제 끝날지 모르는 반복 수행 상황에서 사용
  - 무한루프를 잠재
  - () ~ 무한대
- 반드시 탈출코드가 필요
  - 조건문 + break
  - while 문의 조건식에 변수를 조정
    - 플레그변수(기능)을 이용하여 탈출
      - 탈출용으로 변수를 준비
      - 중복 반복문에서 유용함.
    - 조건식에 걸린 변수값의 변화에 의해서 탈출

- 문법
  ```
    while 조건식:
     statements
  ```

````python
# 더미데이터
datas = [11, 12, 13]

len(datas) # 멤버수 확인
````

````python
datas = [11,12,13]
# datas에 맴버가 없을때까지 반복하여, 각 맴버를 출력하시오,
# 단(맴버를 추출( 리스트.pop() ) 하여 출력함)
# pop() 맴버를 제거하고, 그 값을 반환하는 함수
while len(datas) > 0: #맴버가없을때까지:
  print( datas.pop() )
  pass
````

````python
datas = [11,12,13]
while len(datas):
  print( datas.pop() )
  pass
````

````python
datas = [11,12,13]
while datas: # 가장 간결하게 표현
  print( datas.pop() )
  pass
````
